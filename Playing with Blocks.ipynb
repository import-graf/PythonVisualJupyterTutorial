```python
grid = BlockGrid(5,5)
for x in range(5):
    for y in range(5):
        if y == 1:
            grid[x,y] = (255,0,0)
        elif y == 2:
            grid[x,y] = (0,255,0)
        elif y == 3:
            grid[x,y] = (0,0,255)
        else:
            grid[x,y] = (255,215,0)
        if x == 0 or x == 4:
            grid[x,y] = (0,0,0)   
grid
```

```python
grid = BlockGrid(5,5)
for x in range(5):
    for y in range(5):
        if y == 1:
            grid[x,y] = (255,0,0)
        elif y == 2:
            grid[x,y] = (0,255,0)
        elif y == 3:
            grid[x,y] = (0,0,255)
        else:
            grid[x,y] = (255,215,0)
        if x == 0 or x == 4:
            grid[x,y] = (0,0,0)
        if y == 3 and grid[x,y].rgb == (0,0,0):
            grid[x,y] = (0,0,255)
grid
```
